## Science of Cereal Selection

Parent Rules: can only have "medium" sugary cereals
Kids Goal: Find best (aka junkiest) cereal within parent's rules



```python
median_sugar = df["sugars"].median()
median_sugar
```

### Find Cereals with median sugar

```python
df[df["sugars"] == median_sugar]
```

### Only 4 cereals match median sugar exactly, and none of them feel sugary enough. Frosted mini-wheats???

Let's ask parents if they'll loosen the rules to give more options.

### Make the case that we're TOTALLY not trying to get the most sugary one

```python
sweetest = df.sort_values(by="sugars", ascending=False).iloc[0]
print("We are NOT trying to get the most sugary, which is", sweetest["name"])
```

### How about we sort them by sugars and grab 10 from the middle

```python
mid = len(df) // 2
df.sort_values(by="sugars", ascending=False)[mid-5:mid+5]
```

### Honestly, still not great choices. Cheerios? Not even the Honey Nut kind.

But I doubt parent's will let us go any higher in sugar. 

Just for the heck of it let's show ALL the cereals that are median sugars or less.

```python
df.sort_values(by="sugars", ascending=False)[mid:]
```

### Do my eyes deceive me? There at index 41. It's LIFE !!!!! We're saved

And it's actually low-ish on the sugar scale. We might be able to get convince parent's were emphasizing health for breakfast so maybe we go extra sugary for dessert!

```python
mean_sugars = df["sugars"].mean()
life_sugars = df[df["name"] == "Life"]["sugars"]
diff_sugars = mean_sugars - life_sugars
standard_deviation = df["sugars"].std()
deviations = diff_sugars / standard_deviation
deviations
```

### Dear parents. We elected to NOT choose the most sugary cereal available. We recognize the importance of limiting sugar at breakfast just in case we have a sugary dessert. Therefore we have selected Life cereal as our choice which is a respectable .2 standard deviations below the mean. How does S'Mores sound for dessert?
